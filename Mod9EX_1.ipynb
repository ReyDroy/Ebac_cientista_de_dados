### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# preparando dataframe filtrado
menos_bebe = sinasc_raw[['IDADEMAE','IDADEPAI','munResNome','QTDFILVIVO']]
menos_bebe
```

```python
# municipio que nasceu menos bebe em 2019

menos_bebe.groupby(['munResNome'])['QTDFILVIVO'].count().sort_values(ascending=True)[:5]
```

O município com menos bebes nascidos vivos no ano de 2019 é 'Castanheiras' com 32 bebes. O 'Município ignorado - RO' apresenta uma possível falta de informações, pois é mostrado um resultado de apenas 1 bebe nascido vivo no ano de 2019.

```python
# idade media, maxima, minima das MAES/PAIS nesse municipio
menos_bebe[menos_bebe['munResNome'] == 'Castanheiras'].groupby('munResNome').aggregate(['min', np.median, max]).drop(columns=['QTDFILVIVO'])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
# transformando o DTNASC em datetime

sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])

# criando coluna mês, para facilitar analise

sinasc_raw['MÊS'] = pd.DatetimeIndex(sinasc_raw['DTNASC']).month
```

```python
# preparando dataframe filtrado

mais_bebe = sinasc_raw[['IDADEPAI','munResNome','QTDFILVIVO', 'MÊS']]

mais_bebe.head(12)
```

```python
# municipio que nasceu mais bebe no mês de março

mais_bebe[mais_bebe['MÊS'] == 3].groupby(['munResNome'])['QTDFILVIVO'].count().sort_values(ascending=False)[:5]
```

O município com mais bebes nascidos vivos no ano de 2019 é 'Porto Velho' com 627 bebes.

```python
# quantidade de filhos vivos media, maxima, minima nesse municipio

mais_bebe[mais_bebe['munResNome'] == 'Porto Velho'].groupby('munResNome').aggregate([np.median,max,'min']).drop(columns=['MÊS','IDADEPAI'])
```

```python
# idade media, maxima, minima dos pais nesse municipio

mais_bebe[mais_bebe['munResNome'] == 'Porto Velho'].groupby('munResNome').aggregate([np.median,max,'min']).drop(columns=['MÊS','QTDFILVIVO'])
```
